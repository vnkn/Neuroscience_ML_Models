```python
import model as m
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2

```

```python
sims = []
colors = ['red', 'blue', 'green']
for k in range(100):
    sims.append(k)
def flip(p):
    return 1 if random.random() < p else 0
def run_simulation(target_spike_prob, distractor_spike_prob, nsims):
    for i in range(3):
        response = []
        Mod = m.Model()
        for k in range(nsims):
            f_target = flip(target_spike_prob)
            f_distractor = flip(distractor_spike_prob)

            response.append(Mod.step(f_target, f_distractor))
        plt.plot(sims,response, color = colors[i])
        plt.xlabel('Number of Simulations')
        plt.ylabel('Response')
        plt.title('Response with Target Probability ' + str(target_spike_prob) + ' and Distractor Probability ' + str(distractor_spike_prob) )
    plt.show()
```
